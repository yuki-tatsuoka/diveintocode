# 【問題6】分布の確認
seabornのsns.distplot()を使い目的変数の分布を表示し、「尖度」と「歪度」も算出する。
目的変数に対して対数変換を行う。
対数変換したものに対して分布を表示し、「尖度」と「歪度」も算出する。

対数変換の前後での変化を「正規分布」という言葉を含めて説明してください。このように加工することで予測時の精度を上げることができます。


```python
import numpy as np
from scipy.stats import skew
from scipy.stats import kurtosis

# 目的変数の抽出
SalePrice = df["SalePrice"]

# 歪度と尖度を抽出する
objective_skew1 = skew(SalePrice)
objective_kurtosis1 = kurtosis(SalePrice)
print("尖度{}, 歪度{}".format(objective_skew1, objective_kurtosis1))

#図式化
sns.distplot(SalePrice)
plt.show()

# 対数変換
log_SalePrice = np.log(SalePrice)

# 歪度と尖度を抽出する
objective_skew2 = skew(log_SalePrice)
objective_kurtosis2 = kurtosis(log_SalePrice)
print("尖度{}, 歪度{}".format(objective_skew2, objective_kurtosis2))

# 図式化
sns.distplot(log_SalePrice)
plt.show()

'''
考察
結論から言うと、尖度、歪度共に改善され０付近の値に戻りほとんど正規分布に従う分布になっている。
元のデータでは、正規分布から離れているため正規分布に従うように対数変換を実施する方が好ましい。

また、対数変換は弾力性を見ており、〇〇％増加などの値を見ている。
対数変換を実施しない場合は、通常xの増加量1に対し、yが40増加など実数で見ている。

logの対数はグラフで見ると、0になるまでに急激に増加し〇〇％増加などのサンプル数が大きくなる事により誤差が小さくなる傾向がある。
基本的に歪度があれば対数変換をし、グラフで確認する事が好ましいがyが0以下ならば基本的には誤差が小さくなるので実施した方が良い。
（あとはデータの散らばりも減るので、過学習を抑制しやすいと考えられる）
'''
```
